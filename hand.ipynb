```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

# 1.데이터 불러오기

```python
df= pd.read_csv('data/digit.csv')
df
```

# 2. 데이터에 대한 시각화

```python
sns.countplot(data=df,x='label')
```

```python
df.describe()
```

```python
numbers = df.drop(['label'],axis=1)

nth=3
img=np.reshape(numbers.iloc[nth].values,[28,28])
plt.imshow(img)
plt.show()
```

# 3.데이터셋 나눠주기 (train,valid,test)

```python
train_data=df.drop('label',axis=1)
target_data=df['label']
```

```python
target_data
```

```python
from sklearn.model_selection import train_test_split

#train : test 8:2 나눠주기
x_train,x_test,y_train,y_test=train_test_split(train_data,target_data,test_size=0.2)

#train : valid 8:2 나눠주기
x_train, x_valid, y_train, y_valid= train_test_split(x_train,y_train,test_size=0.2)
```

```python
print('train:',x_train.shape)
print('valid:',x_valid.shape)
print('test:',x_test.shape)
```

# 4. 모델학습하기

### 4-1 DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

#의사결정 나무 , 한가지의 트리만 이용한경우
tree1=DecisionTreeClassifier()
tree1.fit(x_train,y_train)
print('tree1의 train score:',tree1.score(x_train,y_train))
print('tree1의 valid score:',tree1.score(x_valid,y_valid))

```

### 4-2 RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

forest=RandomForestClassifier()
forest.fit(x_train,y_train)

print('ensemble모델의 train score:',forest.score(x_train,y_train))
print('ensemble모델의 valid score:',forest.score(x_valid,y_valid))
```

### 4-3 forest의 파라미터 변경을 통한 정확도 높이기

```python
from sklearn.ensemble import RandomForestClassifier

forest=RandomForestClassifier(max_depth=100,n_estimators=100)
forest.fit(x_train,y_train)

print('ensemble모델의 train score:',forest.score(x_train,y_train))
print('ensemble모델의 valid score:',forest.score(x_valid,y_valid))
```

```python
from sklearn.ensemble import RandomForestClassifier

forest=RandomForestClassifier(max_depth=200,n_estimators=200)
forest.fit(x_train,y_train)

print('ensemble모델의 train score:',forest.score(x_train,y_train))
print('ensemble모델의 valid score:',forest.score(x_valid,y_valid))
```

```python
#정확도가 제일 높은 모델
from sklearn.ensemble import RandomForestClassifier

forest=RandomForestClassifier(max_depth=300,n_estimators=300)
forest.fit(x_train,y_train)

print('ensemble모델의 train score:',forest.score(x_train,y_train))
print('ensemble모델의 valid score:',forest.score(x_valid,y_valid))
```

```python
print(forest.score(x_test,y_test))
```
